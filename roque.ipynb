```python
import cv2
from PIL import Image, ImageOps
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

```

```python
img = cv2.imread('../input/roqueprueba/captura.png')
plt.imshow(img)
plt.show()
```

```python
i0,j0,k0=img.shape
```

```python
h0 = 3456.0
w0 = 5184.0
```

```python
l=(h0-i0)/2
p=(w0-j0)/2
```

```python
l=int(l)
p=int(p)
```

```python
# npad is a tuple of (n_before, n_after) for each dimension
npad = ((l, l), (p, p), (0, 0))
img2 = np.pad(img, pad_width=npad, mode='constant', constant_values=0)
```

```python
plt.imshow(img2)
plt.show()
```

```python
img3 = np.squeeze(img2)
img4 = Image.fromarray(img3, 'RGB')
```

```python
plt.imshow(img4)
plt.show()
```
